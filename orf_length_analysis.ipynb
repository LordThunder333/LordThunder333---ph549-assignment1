```python
from scipy.optimize import *
from scipy.stats import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
file=open('./orf.csv','r')
l=np.array([]) ## Length
freq=np.array([]) ## Frequency
for each in file:
    X=each.split(',')
#     print(X)
    l=np.append(l,int(X[0]))
    freq=np.append(freq,float(X[1]))
file.close()
```

```python
freq=np.log(freq/sum(freq))
```

```python
linregress(l,freq)
```

```python
## y= mx+ a
m=-0.04699327344451327
a=-3.182267256336975
plt.scatter(l,freq,s=0.2,label='Generated')
plt.scatter(l,m*l+a,s=0.2,label='Fit')
plt.xlabel("ORF Length")
plt.ylabel("log(probability)")
plt.legend()
plt.title("Comparision between fit and theoretical data")
plt.savefig('./orf.pdf')
```
